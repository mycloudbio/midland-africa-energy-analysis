```python
# Convert Timestamp to datetime
pd.to_datetime(df["Timestamp"])

# insert into dataframe
df["Timestamp"] = pd.to_datetime(df["Timestamp"])
```

```python
# getting the hours of the timestamp 
df["Timestamp"].dt.hour

# storing it in a variable "hour"
df["Hours"] = df["Timestamp"].dt.hour

df
```

```python
# I'm grouping by "Hours" and getting average "charger power"
df.groupby("Hours")["charger power(W)"].mean()

# putting it ina variable called power voltage hourly (pv_hourly)
pv_hourly = df.groupby("Hours")["charger power(W)"].mean()
pv_hourly
```

```python
# with the use of Matplotlib documentation (https://matplotlib.org/stable/gallery/index.html)
# I have built the Bar Chart to Show US Peak hour of the day when PV system produces the most energy. Using our grouped data


plt.figure(figsize = (10, 6))
pv_hourly.plot(kind="bar")
plt.title("Average pv Charger Power by Hour of the Day")
plt.ylabel("Average Charger Power(W)")
plt.xlabel("Hour of the Day")
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.show()
```

# Peak Generation Analysis Final Documentation, Conclusion, and Suggestion

#### Peak hours are between 9am to 4pm
#### From 7am power generation starts climbing and falls off by 4pm onwards
#### Almost 0 power generation from 5pm onwards
#### Based on this analysis Battery charging optimization should be schedule during peak PV

## Load Pattern Recognition

```python
# Load Pattern Recognition
# Since i'm looking at power load (consumption) behaves across the day
# I'm using the timestamp by groupingby ("HOURS") for demand with power generation, and detecting inefficiencies or unusual spikes.


# Group by hour and calculate average load
load_hourly = df.groupby("Hours")["PLoad(W)"].mean()

# Graphical illustration 
plt.figure(figsize=(10,6))
load_hourly.plot(kind="bar")
plt.title("Average Power Load by Hour of Day")
plt.xlabel("Hour of Day")
plt.ylabel("Average Load Power (W)")
plt.xticks(rotation=0)
plt.grid(axis="y", linestyle="--")
plt.tight_layout()
plt.show()
```

# Load Pattern Recognition Final Documentation, Conclusion, and Suggestion

#### Load demand is quite minimal outside the peak periods which gradually increase from 1pm to 7pm
#### Load power drop drastically around 6pm, quite an ANOMALY 
#### Very low Load power was observed from 8pm onwards, during midnight to early morning hours.
#### High mid-day to evening demand, possibly misaligned with peak solar generation.

# Load Shaving Opportunities

```python
# To look for the load shaving opportunity, we'll continue with 
# Group by "HOURS" and calculate average load, then reset index 
load_hourly_avg = df.groupby("Hours")["PLoad(W)"].mean().reset_index()

# Finding peak load hours
peak_hour = load_hourly_avg.sort_values(by="PLoad(W)", ascending=False).head(8)
peak_hour
```

```python
# Average PV and battery contribution by hour
pv_hourly = df.groupby("Hours")["charger power(W)"].mean().reset_index()
battery_hourly = df.groupby("Hours")["batt power(W)"].mean().reset_index()

# Merge all
shaving_df = load_hourly_avg.merge(pv_hourly, on="Hours").merge(battery_hourly, on="Hours")
shaving_df.columns = ["Hours", "avg_load", "avg_pv", "avg_batt"]
shaving_df

# sort average Load by descending to check what hours of the day has more load spike 
shaving_df.sort_values(by="avg_load", ascending=False).head(8)
```

```python
# Visualizatiion 

plt.figure(figsize=(12,6))
plt.plot(shaving_df["Hours"], shaving_df["avg_load"], label="Avg Load (W)")
plt.plot(shaving_df["Hours"], shaving_df["avg_pv"], label="Avg PV Generation (W)", linestyle="--")
plt.plot(shaving_df["Hours"], shaving_df["avg_batt"], label="Avg Battery Output (W)", linestyle=":")
plt.title("Load vs PV and Battery Contribution")
plt.xlabel("Hour of Day")
plt.ylabel("Power (W)")
plt.legend()
plt.grid(True)
plt.show()
```

# Load Shaving Opportunities Final Documentation, Conclusion, and Suggestion

#### Started off with grouping data by hour of day to find average load across the day, then identified peak demand periods using PLoad(W).
#### Futhermore, i compared load peaks with available PV generation (charger power(W)) and battery power (batt power(W)).
#### The goal was to assess mismatch and cheak if load shifting is neccessary or shaving strategies
#### I observed that Load spikes between 6 PM and 9 PM, battery output is partially supporting load also
#### Implement smart load control systems, and Increase battery storage to cover peak periods might be something we should consider

# System Optimization Timing

```python
# I continued with using the Hours timestamp, and placed it in a variable called optimized_df
# and calculated the mean of charger power(W); batt power(W); PLoad(W)

optimized_df = df.groupby("Hours")[['charger power(W)', 'batt power(W)', 'PLoad(W)']].mean().round(2)
optimized_df
```

```python
# Visualizatio 


plt.figure(figsize=(12, 6))
plt.plot(optimized_df.index, optimized_df["charger power(W)"], marker="o", label="Charger Power (W)", color="green")
plt.plot(optimized_df.index, optimized_df["batt power(W)"], marker="o", label="Battery Power (W)", color="orange")
plt.plot(optimized_df.index, optimized_df["PLoad(W)"], marker="o", label="Load Power (W)", color='blue')
plt.title("System Optimization Timing - Mean Power by Hour")
plt.xlabel("Hour of Day")
plt.ylabel("Power (W)")
plt.xticks(range(0, 24))
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
